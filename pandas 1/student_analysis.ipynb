## Clean the data up

```python
df.info()
```

```python
df['writing score'] = pd.to_numeric(df['writing score'])
df['math score'] = pd.to_numeric(df['math score'])
```

```python
df.info()
```

```python
# do we have missing values?

df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
# df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.info()
df.shape
```

```python
# compute average scores

df['average_score'] = (df['writing score'] + df['reading score'] + df['math score'])/3
```

```python
df.head(5)
```

```python
df.columns
```

```python
new_columns = []
for col in df.columns:
    col = col.replace("/", "_").replace(' ', '_')
    new_columns.append(col)
print(new_columns)
```

```python
# let's assign them to the dataframe

df.columns = new_columns
```

```python
df
```

```python
# Let's assign the grades

# create a new column for the grades

# use numpy
conditions = [(df.average_score >= 90),
              (df.average_score <90) & (df.average_score >= 80),
              (df.average_score < 80)
]

values = ['A', 'B', 'Pass']

# using a function

df['grades'] = np.select(conditions, values)
```

```python
df
```

```python
# using apply
# define the function
def compute_grade(row):
    # you would do your math here etc
    if row.average_score >= 90:
        return 'A'
    elif row.average_score <90 and row.average_score >= 80:
        return 'B'
    else:
        return 'Pass'
```

```python
# now apply the function
df['grade_2'] = df.apply(compute_grade, axis= 1)
```
